## Numpy

Fonctions utilisées dans ce notebook:

* [`arange([start=0],stop,[step=1],[dtype=None])`](https://numpy.org/doc/stable/reference/generated/numpy.arange.html): Retourne des valeurs uniformément espacées dans l'intervalle donné.
* [`linespace(start,stop,[num=50],[endpoint=True],[retstep=False],[dtype=None],[axis=0])`](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html): Retourne des nombres régulièrement espacés dans l'intervalle donné.
* [`ones(shape,[dtype=None],[order='C'])`](https://numpy.org/doc/1.19/reference/generated/numpy.ones.html): Retourne un tableau de forme donné, par exemple shape=4 ou shape=(3,4), rempli de 1.
* [`outer(u,v,[dtype=None])`](): Retourne le produit dyadique $p=u \otimes v$ de deux vecteurs.
* [`sin(x)`](https://numpy.org/doc/1.19/reference/generated/numpy.sin.html): Retourne une table de la même forme que `x` contenant le sinus par element de `x` (radians)
* [`cos(x)`](https://numpy.org/doc/1.19/reference/generated/numpy.cos.html): Retourne une table de la même forme que `x` contenant le cosinus par element de `x` (radians)
* [`sqrt(x)`](https://numpy.org/doc/1.19/reference/generated/numpy.sqrt.html): Retourne une table de la même forme que `x` contenant la racine carrée par element de `x`
* [`meshgrid(*x,[copy=True],[sparse=False],[indexing='xy'])`](https://numpy.org/doc/1.19/reference/generated/numpy.meshgrid.html): Retourne un tableau de dimensions N-D, étant donné $n$ vecteurs $x_1, x_2, x_n$.

### Diagramme simple

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

#fig.savefig("data/test1.png")
plt.show()
```

### Bar plots

```python
import numpy as np
import matplotlib.pyplot as plt


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std, bottom=men_means,
       label='Women')

ax.set_ylabel('Ages')
ax.set_title('Ages by group and gender')
ax.legend()

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ages')
ax.set_title('Ages by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```

### Diagramme circulaire

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Isoplèthes

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
fig, ax=plt.subplots()
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
fig, ax = plt.subplots()
cp = ax.contour(X, Y, Z)
ax.clabel(cp, inline=1, fontsize=10)
ax.set_title('Contour Plot with labels')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```

### Quiver plot

```python
import matplotlib.pyplot as plt
import numpy as np
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2)
ax=plt.axes()
q = ax.quiver(x,y,u,v)
plt.show()
```

### Courbe parametrique 3D

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['legend.fontsize'] = 10

fig = plt.figure()
#ax = plt.axes(projection='3d')
#ax = fig.add_subplot(111,projection='3d')
ax = fig.gca(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

### Diagrammes de surfaces 3D

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.outer(np.linspace(-1.5, 1.5, 30), np.ones(30))
y = x.copy().T # transpose
z = np.cos(x ** 2 + y ** 2)
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot $\cos{(x^2 + y^2)}$')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = plt.axes(projection='3d')

# Make data
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the surface
ax.plot_surface(x, y, z)

plt.show()
```
